Nos encontramos ante un problema de clasificación, puesto que el objetivo del conjunto de datos y del modelo a crear es predecir si un empleado puede abandonar la empresa o no, por lo que el modelo se encargará de clasificar a los distintos empleados en si pueden abandonar la empresa o no. La variable de salida es attrition, la cual tiene dos posibles valores, 'Yes' o 'No', por lo que estamos ante un problema de clasificación binaria. Como puede observarse, el problema es desbalanceado, encontrando que un 83,87% de las instancias pertenecen a la clase 'No' (attrition = 'No'), por lo que significa que un modelo trivial tendría un acierto del 83,87% si clasificara todos los datos como 'No'.

# **Train/Test Split**

---



```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline

# Procesamiento de valores numéricos.
features_num = ["hrs", "absences", "JobInvolvement", "PerformanceRating", "EnvironmentSatisfaction", "JobSatisfaction", "WorkLifeBalance", "Age", "DistanceFromHome", "Education","EmployeeID", "JobLevel","MonthlyIncome", "NumCompaniesWorked", "PercentSalaryHike", "StockOptionLevel", "TotalWorkingYears", "TrainingTimesLastYear", "YearsAtCompany", "YearsSinceLastPromotion","YearsWithCurrManager"]
imputer_num = SimpleImputer(strategy='mean')
pipeline_num = Pipeline(
    steps=[
        ("imputer", imputer_num)
    ]
)

# Procesamiento de valores categóricos.
features_cat = ["BusinessTravel", "Department", "EducationField", "Gender", "JobRole", "MaritalStatus"]
imputer_cat = SimpleImputer(strategy='most_frequent')
pipeline_cat = Pipeline(
    steps=[
        ("imputer", imputer_cat)
    ]
)

# Pre-procesador "global".
# Dependiendo del tipo de columna se aplica una transformación u otra.

processor = ColumnTransformer(
    transformers=[
        ("num", pipeline_num, features_num),
        ("cat", pipeline_cat, features_cat),
    ]
)

# Realizamos la transformación.
disp_df_final = processor.fit_transform(disp_df.drop(columns='Attrition'))
```

```python
#Comprobamos el nuevo dataset obtenido
disp_df = pd.DataFrame(disp_df_final, columns=["hrs", "absences", "JobInvolvement", "PerformanceRating", "EnvironmentSatisfaction", "JobSatisfaction", "WorkLifeBalance", "Age", "DistanceFromHome", "Education","EmployeeID", "JobLevel","MonthlyIncome", "NumCompaniesWorked", "PercentSalaryHike", "StockOptionLevel", "TotalWorkingYears", "TrainingTimesLastYear", "YearsAtCompany", "YearsSinceLastPromotion","YearsWithCurrManager","BusinessTravel", "Department", "EducationField", "Gender", "JobRole", "MaritalStatus"])

print(disp_df)
print(type(disp_df))
```

Aquellos atributos cuyos valores son del tipo *object* debrán ser codificados usando valores numericos. Para ello se utilizará one-hot-encoding con aquellos que no cumplan con los requisitos para una codificación ordinaria. En este caso, solo *BusinessTravel* podrá ser codificado de esa manera por lo que *Department, EducationField, Gender, JobRole y MaritalStatus* se codificicrán de esa forma.

```python
#Obtenemos datasets temporales mde cada atributo codificados mediante one-hot-encoding con get_dummies()
df_temp1=pd.get_dummies(disp_df.Department)
df_temp2=pd.get_dummies(disp_df.EducationField)
df_temp3=pd.get_dummies(disp_df.Gender)
df_temp4=pd.get_dummies(disp_df.JobRole)
df_temp5=pd.get_dummies(disp_df.MaritalStatus)

#Eliminamos las columnas antiguos de dichos atributos y añadimos las nuevas
disp_df=disp_df.drop(columns=['Department', 'EducationField', 'Gender','JobRole','MaritalStatus'])
df_ohe=[disp_df,df_temp1,df_temp2, df_temp3, df_temp4, df_temp5]

disp_df=pd.concat(df_ohe, axis=1)
disp_df
```

```python
#Codificacmos temporalmente un nuevo dataset con la codificación ordinaria de BusinessTravel
df_temp6=disp_df.copy()

scale_mapper = {"Non-Travel":1, "Travel_Rarely":2, "Travel_Frequently":3}

df_temp6=df_temp6.drop(columns=["hrs", "absences", "JobInvolvement", "PerformanceRating", "EnvironmentSatisfaction", "JobSatisfaction", "WorkLifeBalance", "Age", "DistanceFromHome", "Education","EmployeeID", "JobLevel","MonthlyIncome", "NumCompaniesWorked", "PercentSalaryHike", "StockOptionLevel", "TotalWorkingYears", "TrainingTimesLastYear", "YearsAtCompany", "YearsSinceLastPromotion","YearsWithCurrManager"])

df_temp6["BusinessTravel"] = df_temp6["BusinessTravel"].replace(scale_mapper)

df_temp6=df_temp6.BusinessTravel
df_temp6

#Eliminamos las columnas antiguos de dichos atributos y añadimos las nuevas
disp_df=disp_df.drop(columns=["BusinessTravel"])

df_final_array=[disp_df,df_temp6]

disp_df=pd.concat(df_final_array, axis=1)
disp_df
```

```python
#Codificacmos el valor de clasificación con valores numéricos
scale_mapper = {"No":0, "Yes":1}

disp_df_target = disp_df_target.replace(scale_mapper)

disp_df_target
```

```python
#Dividimos los datos en train y test
data=disp_df.to_numpy()
target=disp_df_target.to_numpy()

print(data)
print(target)

print("Tamaños de los arrays: ")
print("Global: " + str(disp_df.shape))
print("Entradas: " + str(data.shape))
print("Salidas: " + str(target.shape))
```

```python
#Realizamos el train/test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.33, random_state=42, stratify=target)
```

```python
#Cargamos e importamos las librerías necesarias
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import balanced_accuracy_score, f1_score, confusion_matrix
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

#Creamos el clasificador y la pipeline
pipeline = make_pipeline(StandardScaler(), LogisticRegression(random_state=42, class_weight='balanced'))

#Entrenamos el modelo y generamos las predicciones. Calculamos las medida de evaluación (Balanced Accurancy, F1-Score y Matriz de confusión)
clf = pipeline.fit(X_train, y_train)

y_pred = clf.predict(X_test)

print("Balanced Accurancy Score: " + str(balanced_accuracy_score(y_test, y_pred)))
print("F1 Score: " + str(f1_score(y_test, y_pred, average='weighted')))
print("Matriz de confusión: ")
print(confusion_matrix(y_test, y_pred))
```

## MODELOS SKLEARN:

### *Random Forest:*

```python
#Cargamos e importamos las librerías necesarias
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold

#Elaboramos el espacio de búsqueda de hiperparámetros
param_grid = {
    'max_depth': range(10, 50, 2),
    'n_estimators': range(50, 200, 50),

}

#Creamos el clasificador con el ajuste de parámetros automático
Random_forest_classifier = GridSearchCV(RandomForestClassifier(), 
                    param_grid,
                    scoring='f1',
                    verbose=1,
                    cv=StratifiedKFold(n_splits=5, random_state=None, shuffle=False),
                    refit=False)

clf_Rand_Forest = make_pipeline(StandardScaler(), Random_forest_classifier)
np.random.seed(42)

#Entrenamos el modelo y generamos las predicciones. Calculamos las medida de evaluación (Balanced Accurancy, F1-Score y Matriz de confusión)
clf_Rand_Forest.fit(X_train, y_train)
clf_Rand_Forest_Best = RandomForestClassifier(**Random_forest_classifier.best_params_)
clf_Rand_Forest_Best.fit(X_train, y_train)
y_pred = clf_Rand_Forest_Best.predict(X_test)

print("Balanced Accurancy Score: " + str(balanced_accuracy_score(y_test, y_pred)))
print("F1 Score: " + str(f1_score(y_test, y_pred, average='weighted')))
print("Matriz de confusión: ")
print("Class'No'  | Class 'Yes' ")
print(confusion_matrix(y_test, y_pred))
```
