Nos encontramos ante un problema de clasificación, puesto que el objetivo del conjunto de datos y del modelo a crear es predecir si un empleado puede abandonar la empresa o no, por lo que el modelo se encargará de clasificar a los distintos empleados en si pueden abandonar la empresa o no. La variable de salida es attrition, la cual tiene dos posibles valores, 'Yes' o 'No', por lo que estamos ante un problema de clasificación binaria. Como puede observarse, el problema es desbalanceado, encontrando que un 83,87% de las instancias pertenecen a la clase 'No' (attrition = 'No'), por lo que significa que un modelo trivial tendría un acierto del 83,87% si clasificara todos los datos como 'No'.

# **Train/Test Split**

---



```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline

# Procesamiento de valores numéricos.
features_num = ["hrs", "absences", "JobInvolvement", "PerformanceRating", "EnvironmentSatisfaction", "JobSatisfaction", "WorkLifeBalance", "Age", "DistanceFromHome", "Education", "EmployeeCount", "MonthlyIncome", "NumCompaniesWorked", "PercentSalaryHike", "StandardHours", "StockOptionLevel", "TotalWorkingYears", "TrainingTimesLastYear", "YearsAtCompany", "YearsSinceLastPromotion"]
imputer_num = SimpleImputer(strategy='mean')
pipeline_num = Pipeline(
    steps=[
        ("imputer", imputer_num)
    ]
)

# Procesamiento de valores categóricos.
features_cat = ["BusinessTravel", "Department", "EducationField", "Gender", "JobRole", "MaritalStatus", "Over18"]
imputer_cat = SimpleImputer(strategy='most_frequent')
pipeline_cat = Pipeline(
    steps=[
        ("imputer", imputer_cat)
    ]
)

# Pre-procesador "global".
# Dependiendo del tipo de columna se aplica una transformación u otra.

processor = ColumnTransformer(
    transformers=[
        ("num", pipeline_num, features_num),
        ("cat", pipeline_cat, features_cat),
    ]
)

# Realizamos la transformación.
disp_df_final = processor.fit_transform(disp_df.drop(columns='Attrition'))
```

```python
y = pd.get_dummies(disp_df['BusinessTravel'], prefix='Travel')
print(y.head())
```

```python
data=disp_df_final
target=disp_df['Attrition'].to_numpy()

print(data)
print(target)

print("Tamaños de los arrays: ")
print("Global: " + str(disp_df.shape))
print("Entradas: " + str(data.shape))
print("Salidas: " + str(target.shape))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.33, random_state=42, stratify=target)
```

```python
unique_train, counts_train = np.unique(y_train, return_counts=True)
unique_test, counts_test = np.unique(y_test, return_counts=True)
print(dict(zip(unique_train, counts_train)))
print(dict(zip(unique_test, counts_test)))
print("Porcentaje de datos de la clase NO en:")
print("Train: " + str(counts_train[0]/y_train.shape[0]))
print("Test: " + str(counts_test[0]/y_test.shape[0]))
print("Porcentaje de datos de la clase YES en:")
print("Train: " + str(counts_train[1]/y_train.shape[0]))
print("Test: " + str(counts_test[1]/y_test.shape[0]))
```

# ***Modelo*** ***LogisticRegression***

---



```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import balanced_accuracy_score, f1_score, confusion_matrix

clf = LogisticRegression(random_state=42, class_weight='balanced').fit(X_train, y_train)

y_pred = clf.predict(X_train)

balanced_accuracy_score(y_test, y_pred)
f1_score(y_test, y_pred, average='weighted')
confusion_matrix(y_test, y_pred)
```
